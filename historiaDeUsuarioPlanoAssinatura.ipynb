# Buscando likes e dislikes recebidos nas mensagens

```sql
SELECT 
       mensagem.id, 
       mensagem.mensagem,
       consumidor.nome as avaliacaocriadapor,
       (SELECT COUNT(avaliacao.gostou) 
        FROM avaliacao
        WHERE avaliacao.gostou = 1 AND avaliacao.mensagemid = mensagem.id) AS gostou,
        (SELECT COUNT(avaliacao.gostou) 
        FROM avaliacao
        WHERE avaliacao.gostou = 0 AND avaliacao.mensagemid = mensagem.id) AS naogostou
FROM forumassinatura
LEFT JOIN mensagem ON mensagem.forumid = forumassinatura.id
INNER JOIN consumidor ON mensagem.consumidorid = consumidor.id;
```

# Buscando avaliacao de um plano, quem criou quem respondeu e a mensagem da resposta referente a avaliação

```sql
SELECT forumassinatura.id,
       forumassinatura.titulo,
       mensagem.mensagem,
       consumidor.nome as avaliacaocriadapor,
       consumidoravaliacao.nome as consumidoresposta,
       avaliacao.descricaoavaliacao as resposta
FROM forumassinatura
LEFT JOIN mensagem ON mensagem.forumid = forumassinatura.id
INNER JOIN consumidor ON mensagem.consumidorid = consumidor.id
LEFT JOIN avaliacao ON mensagem.id = avaliacao.mensagemid
INNER JOIN consumidor consumidoravaliacao ON avaliacao.consumidorid = consumidoravaliacao.id;
```

# Buscando planos de assinatura para comparação

```sql
SELECT nomeexibicao, descricao FROM planoassinatura;
```

# Obtendo plano de assinatura de um usuário

```sql
SET @idAssinante = 1;

SELECT consumidor.nome,
       assinante.datamensalidade,
       planoassinatura.nomeexibicao,
       planoassinatura.descricao
FROM consumidor
INNER JOIN assinante ON assinante.consumidorid = consumidor.id
INNER JOIN planoassinatura ON assinante.planoid = planoassinatura.id
WHERE assinante.id = @idAssinante;       
```

# Obtendo os planos com mais assinatura para um possível dashboard

```sql
SELECT planoassinatura.nomeexibicao, COUNT(*) as quantidade
FROM assinante
INNER JOIN planoassinatura ON planoassinatura.id = assinante.planoid
GROUP BY 1
ORDER BY quantidade desc;
```
